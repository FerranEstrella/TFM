```python
#!pip install equinox optax diffrax
import jax
import jax.numpy as jnp
import jax.nn as jnn
import jax.random as jrandom
from jax import vmap

import equinox as eqx  # JAX library
import optax # JAX optimizer library
import diffrax as dfx # Optimized numerical integration

import matplotlib.pyplot as plt
import os
from tqdm import tqdm

# 1. HYPERPARAMETERS AND KNOWN FEATURES

# DeepONet hyperparameters
basis_size = 64
system_dimension=2
hidden_dim = 64
latent_dim = 128
depth = 4

# Fourier Features
F = [0.1, 0.25, 0.5, 1.0, 2.0, 3.0]
N_Fourier = 2*len(F)

# Period and Floquet exponent for the considered system:
T = 21.36555175500626
lamb = -0.15358058186323872

# Explored range of parameters and IC
mu_dim = 1
eps_min = 0.0
eps_max = 1.0
x0_dim = 2
ic_min = jnp.array([0.0, -1.0])
ic_max = jnp.array([T/4, 1.0])

# Explored time interval
t_min = 0.0
t_max = T

# Supervised data
N_traj = 1600 # Number of supervised trajectories, that is, considered number of active parameter and IC (in the considered range)
N_time = 128 # Number of timesteps (in the considered range)
dt0 = 1e-3 # Integration timestep

# Training
learning_rate = 1e-4
Batch_size = 128
N_sup = 128   # Number of supervised samples (from precomputed data)
N_coll = 0   # Number of collocation samples (freshly sampled)
N_epoch = 25000

# 2. DEEPONET SETUP

# Function for the baseline solution
def baseline_solution(x0, t):

    theta0, sigma0 = x0[0], x0[1]
    theta_base = theta0 + t
    sigma_base = sigma0 * jnp.exp(lamb * t)

    return jnp.array([theta_base, sigma_base])


# Function for obtaining Fourier features from time input
def fourier_features(t):
    t = jnp.atleast_1d(t)
    freqs = jnp.array(F)
    angles = 2 * jnp.pi * freqs * t
    return jnp.concatenate([jnp.sin(angles), jnp.cos(angles)])


# DeepONet architecture
class MultiInputDeepONet(eqx.Module):
    # Deep Neural sub-Networks
    branch_mu: eqx.nn.MLP
    branch_x0: eqx.nn.MLP
    trunk_t: eqx.nn.MLP
    merger: eqx.nn.MLP

    # Inputs
    mu_active_idx: jnp.ndarray = eqx.field(static=True) # Index of the active params in training
    system_dimension: int = eqx.field(static=True) # Dimension of the ODE

    # Basis size 'p' for the branch-trunk approximation
    basis_size: int = eqx.field(static=True)

    # Scaling ctts
    t_max: float = eqx.field(static=True)
    mu_min: jnp.ndarray = eqx.field(static=True)
    mu_max: jnp.ndarray = eqx.field(static=True)
    u_scale: jnp.ndarray = eqx.field(static=True)

    # DeepONet variables
    def __init__(self, mu_dim, x0_dim, latent_dim, basis_size,
                 system_dimension, t_max, mu_active_idx,
                 mu_min, mu_max, u_scale, key):

        keys = jax.random.split(key, 4)

        self.mu_active_idx = jnp.array(mu_active_idx)
        active_dim = len(mu_active_idx)
        self.system_dimension = system_dimension

        self.basis_size = basis_size

        self.t_max = t_max
        self.mu_min = jnp.array(mu_min)
        self.mu_max = jnp.array(mu_max)
        self.u_scale = jnp.array(u_scale)


        # Branch mu: Extracts features from parameters
        self.branch_mu = eqx.nn.MLP(active_dim, latent_dim, width_size=hidden_dim, depth=depth,
                                   activation=jax.nn.tanh, key=keys[0])

        # Branch x0: Extracts features from initial conditions
        self.branch_x0 = eqx.nn.MLP(x0_dim, latent_dim, width_size=hidden_dim, depth=depth,
                                   activation=jax.nn.tanh, key=keys[1])

        # Merger: Combines the two latent vectors from mu and x0 branches
        self.merger = eqx.nn.MLP(
            latent_dim * 2, system_dimension * basis_size, width_size=hidden_dim, depth=depth, # Input size is latent_dim * 2 because of concatenation
                                    activation=jax.nn.tanh, key=keys[2])

        # Trunk: Learns the temporal basis functions
        self.trunk_t = eqx.nn.MLP(N_Fourier, basis_size, width_size=hidden_dim, depth=depth, # Input size is the number of Fourier features
                                    activation=jax.nn.tanh, key=keys[3])

    # DeepONet prediction
    def __call__(self, mu, x0, t):

        mu_active = mu[self.mu_active_idx] # Only consider active parameters in training
        mu_norm = 2.0 * (mu_active - self.mu_min) / (self.mu_max - self.mu_min) - 1.0   # Map parameters to [-1, 1] range for better Tanh performance

        t_norm = t / self.t_max # Time normalization
        t_feat = fourier_features(t_norm) # Extract time features
        #t_feat = jnp.atleast_1d(t_norm) # Linear time input

        # 0. Baseline
        u_base = baseline_solution(x0, t)

        # 1. Get Coefficients (Branch)
        c = self.merger(jnp.concatenate([
            self.branch_mu(mu_norm),
            self.branch_x0(x0)]))

        # 2. Get Basis Functions (Trunk)
        phi = self.trunk_t(t_feat)

        # 3. Precision Reshape: (d, K)
        c = c.reshape(self.system_dimension, self.basis_size)

        # 4. Matrix-Vector product: (d, K) @ (K,) -> (d,)
        raw_out = c @ phi

        # 5. Calculate the scaled neural network output (rescaling)
        nn_residual = raw_out * self.u_scale

        # 6. Give output considering baseline solution perturbation and IC restriction
        return u_base + mu[0] * (1.0 - jnp.exp(-t_norm)) * nn_residual  # eps = mu[0] scaling to match the eps = 0 case


# ODE of the considered IVP
def f_model(u, mu, W):

    theta, sigma = u[0], u[1]  # Variables of the system
    eps = mu[0]

    # Note: We truncate the Taylor expansions of each function at first order

    Z0 = (
        0.8434178848824566
        + 0.22829012252643252 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        - 0.9917385570922767 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        - 0.11048654332929672 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        - 0.08565806523890455 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        + 0.006828170568351597 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        + 0.03367035578172542 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        - 0.007250339682290018 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        + 0.026051043938219005 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        - 0.012964104203281876 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        + 0.00832549409774979 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        - 0.007420480760365434 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        - 0.0018985283814105221 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        - 0.0013866114449639746 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.003061722187489108 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        + 0.0006618238805303856 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.0010194093249861701 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.00026492611275421 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        + 0.00011376941388258679 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        - 0.00018247661750945394 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        + 8.91073953039363e-6 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    Z1 = (
        -0.032833135435209646
        - 0.008198125371330574 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        + 0.08609884041939658 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        - 0.04133341781850866 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        + 0.007258236171156525 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        - 0.021792286068914292 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        - 0.00021691596432540269 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        - 0.011220138810296564 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        - 0.0038822903858640994 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        - 0.005906558352748338 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        - 0.004820976844058482 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        - 0.0018579634274482647 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        - 0.004607748945557491 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        + 0.0007320832129090896 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.0029215120616025827 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        + 0.0013249244075619784 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.0009313448488039927 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.0007108552149020645 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        + 0.00010682457508628284 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        + 0.00013014347555994764 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        + 0.00015700271217225872 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    I0 = (
        0.0783581580561428
        + 1.5342479515842966 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        - 0.6999763722929998 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        + 0.10859515367994853 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        - 0.5568005818472679 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        + 0.0603380612432804 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        - 0.01695397352559378 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        + 0.017773530860307668 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        + 0.030389681139791826 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        - 0.009276193031855624 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        + 0.01663993871007951 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        - 0.009419938012717522 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        + 0.0005838059855889642 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        - 0.002311117044039185 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.003494617763941366 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        + 0.0009189269568954324 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.0013040747404790918 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.00042594884728826504 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        + 0.0003633586230360605 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        - 0.00036416019700064267 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        + 0.00017143637451412524 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    I1 = (
        -0.04784617626264868
        - 0.7276650745107754 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        + 0.25897335500365043 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        - 0.12684800351057948 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        + 0.12279794029810664 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        - 0.034191167806249166 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        - 0.007980225240505375 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        - 0.005182256855406266 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        - 0.012011936087715112 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        + 0.0006693345604175645 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        - 0.004310606604092428 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        + 6.585876992163817e-8 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        - 0.0015350705789912763 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        + 2.8352178068040994e-5 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.0013819631945833786 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        + 0.0006650644400149661 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.0008544624854824495 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.0006240821372215004 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        - 4.8674778271926284e-5 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        + 0.0001540347402865894 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        + 0.00018856881423242764 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    Kse0 = (
        0.05910068113134753
        + 0.04895999908691315 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        + 0.04810922802634314 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        + 0.0017719704723411103 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        + 0.03308604470775286 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        - 0.007115343568536075 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        + 0.014528916852122327 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        - 0.007249528102163706 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        + 0.0056858297803530355 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        - 0.005447927872907616 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        + 0.0007635856720751247 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        - 0.0028572801829345826 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        - 0.0012819910473445262 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        - 0.000976399885539279 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.0013752562969088873 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        - 0.00013924446447169565 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.0008819302477050103 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.00011708919842875278 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        - 0.0004808489641861803 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        + 0.00017532784010734073 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        - 0.00023850395124322372 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    Kse1 = (
        -0.0030578820899022155
        + 0.007619814159585614 * jnp.cos(2 * jnp.pi * 1 / T * theta)
        + 0.0077206162686050805 * jnp.sin(2 * jnp.pi * 1 / T * theta)
        - 0.0038037030123124864 * jnp.cos(2 * jnp.pi * 2 / T * theta)
        + 0.015241934435253436 * jnp.sin(2 * jnp.pi * 2 / T * theta)
        - 0.01083284059673264 * jnp.cos(2 * jnp.pi * 3 / T * theta)
        + 0.006559966344655594 * jnp.sin(2 * jnp.pi * 3 / T * theta)
        - 0.008641456616118228 * jnp.cos(2 * jnp.pi * 4 / T * theta)
        - 0.0006209646294280072 * jnp.sin(2 * jnp.pi * 4 / T * theta)
        - 0.004753572488192002 * jnp.cos(2 * jnp.pi * 5 / T * theta)
        - 0.0032368338388349136 * jnp.sin(2 * jnp.pi * 5 / T * theta)
        - 0.001738657728014593 * jnp.cos(2 * jnp.pi * 6 / T * theta)
        - 0.0033870381518263755 * jnp.sin(2 * jnp.pi * 6 / T * theta)
        + 4.651274805790124e-5 * jnp.cos(2 * jnp.pi * 7 / T * theta)
        - 0.0024812190377436613 * jnp.sin(2 * jnp.pi * 7 / T * theta)
        + 0.0007565750778070334 * jnp.cos(2 * jnp.pi * 8 / T * theta)
        - 0.001388986707222854 * jnp.sin(2 * jnp.pi * 8 / T * theta)
        + 0.0008127341672505832 * jnp.cos(2 * jnp.pi * 9 / T * theta)
        - 0.000574903362554021 * jnp.sin(2 * jnp.pi * 9 / T * theta)
        + 0.000619902275276155 * jnp.cos(2 * jnp.pi * 10 / T * theta)
        - 0.00011325513799811853 * jnp.sin(2 * jnp.pi * 10 / T * theta)
    )

    dthetadt = 1 + eps * (Z0 + sigma * Z1) * (Kse0 + sigma * Kse1)
    dsigmadt = lamb * sigma + eps * (I0 + sigma * I1) * (Kse0 + sigma * Kse1)


    return jnp.array([dthetadt, dsigmadt])


# Total loss
def loss(model, mu, x0, t, u_true, W, weights, lambda_ode):

    def loss_single(mu, x0, t, u_true, w):
        # Model for fixed params and IC
        def model_t(tt):
            return model(mu, x0, tt)

        # Prediction by the model
        u_pred = jax.vmap(model_t)(t)

        #1) Supervision error
        err = (u_pred - u_true) / model.u_scale
        loss_theta = jnp.mean(err[:, 0]**2)
        loss_sigma = jnp.mean(err[:, 1]**2)

        loss_data = w * (loss_theta + 10.0 * loss_sigma) # Manually weight sigma loss higher. This forces the optimizer to "care" about the sigma gap
                                                         # even if the total loss value is already small (due to theta fit)

        #2) IC error
        #loss_ic = jnp.mean( ((model_t(0.0) - x0) / model.u_scale)**2 )

        #3) ODE error
        # ODE pointwise residual for fixed params and IC
        def res(tt):
            u, du = jax.jvp(model_t, (tt,), (1.0,)) # Jacobian-Vector product, returns prediction + time derivative (jacobian vector produt with time direction)
            physical_residual = du - f_model(u, mu, W)
            return physical_residual * (model.t_max / model.u_scale) # adimensionalization

        loss_ode = jnp.mean(jax.vmap(res)(t)**2)

        #4) Var eq error
        def var_res(tt):
            phi = jax.jacobian(lambda x: model(mu, x, tt))(x0) #Phi = du(t)/du0
            _, dphi_dt = jax.jvp(lambda t_eval: jax.jacobian(lambda x: model(mu, x, t_eval))(x0), (tt,), (1.0,)) #d/dt(Phi)
            u_pred = model(mu, x0, tt)
            Df = jax.jacobian(lambda u_in: f_model(u_in, mu, W))(u_pred) #Df = df/du evaluated at current predicted state
            physical_var_res = dphi_dt - (Df @ phi)
            return physical_var_res # / model.u_scale[:, jnp.newaxis]

        #loss_var = jnp.mean(jax.vmap(var_res)(t)**2)

        return (loss_data + lambda_ode * loss_ode) #, loss_ic, loss_var] --> WE ONLY CONSIDER SUPERVISION ERROR AND, POSSIBLY, ODE ERROR

    return jnp.mean(jax.vmap(loss_single)(mu, x0, t, u_true, weights)) # Average loss for a certain batch


# Parallel computation of the loss (debug)
def loss_debug(model, mu, x0, t, u_true, W, weights, lambda_ode):

    def loss_single_debug(mu, x0, t, u_true, w):
        def model_t(tt):
            return model(mu, x0, tt)

        u_pred = jax.vmap(model_t)(t)

        #1) Supervision error
        err = (u_pred - u_true) / model.u_scale
        loss_theta = jnp.mean(err[:, 0]**2)
        loss_sigma = jnp.mean(err[:, 1]**2)
        loss_data = w * (loss_theta + 10.0 * loss_sigma)

        #2) IC error
        #loss_ic = jnp.mean( ((model_t(0.0) - x0) / model.u_scale)**2 )

        #3) ODE error
        def res(tt):
            u, du = jax.jvp(model_t, (tt,), (1.0,))
            physical_residual = du - f_model(u, mu, W)
            return physical_residual * (model.t_max / model.u_scale)

        loss_ode = jnp.mean(jax.vmap(res)(t)**2)

        #4) Var eq error
        def var_res(tt):
            phi = jax.jacobian(lambda x: model(mu, x, tt))(x0)
            _, dphi_dt = jax.jvp(lambda t_eval: jax.jacobian(lambda x: model(mu, x, t_eval))(x0), (tt,), (1.0,))
            dphi_dt = dphi_dt / model.t_max
            u_pred = model(mu, x0, tt)
            Df = jax.jacobian(lambda u_in: f_model(u_in, mu, W))(u_pred)
            physical_var_res = dphi_dt - (Df @ phi)
            return physical_var_res / model.u_scale[:, jnp.newaxis]

        #loss_var = jnp.mean(jax.vmap(var_res)(t)**2)
        return (loss_data + lambda_ode * loss_ode, loss_data, lambda_ode * loss_ode, loss_ode) #, loss_var]

    #return loss_single_debug( mu[0], x0[0], t[0], u_true[0], weights[0])
    losses = jax.vmap(loss_single_debug)(mu, x0, t, u_true, weights)
    return tuple(jnp.mean(l) for l in losses) # Returns averaged (total, data, ic, ode)


#2. DATA GENERATION

# ODE to be time integrated (diffrax) function
def f_diffrax(t, u, args):
    mu, W = args
    return f_model(u, mu, W)

# Integrator (JAX) function
def integrate(mu, x0, t, W):

    t = jnp.sort(t)
    sol = dfx.diffeqsolve(
        dfx.ODETerm(f_diffrax),
        dfx.Tsit5(),
        t0=t[0],
        t1=t[-1],
        dt0=dt0,
        y0=x0,
        args=(mu, W),
        saveat=dfx.SaveAt(ts=t),
        stepsize_controller=dfx.PIDController(rtol=1e-5, atol=1e-5)
    )

    return sol.ys

# Batch sample: params and x0 (function)
def sample_batch(key, B):
    k1, k2 = jax.random.split(key)
    mu = jnp.zeros((B, 1)) # Single parameter: eps
    mu = mu.at[:, 0].set(jax.random.uniform(k1, (B,), minval=eps_min, maxval=eps_max)) # eps range
    x0 = jax.random.uniform(k2, (B, 2), minval=ic_min, maxval=ic_max) # ic range
    return mu, x0

# Batch sample: time (function)
def sample_times(key, B, Nt): # Nt is the number of timesteps per batch
    k1, k2 = jax.random.split(key)
    t = jax.random.uniform(k1, (B, Nt), minval=t_min, maxval=t_max)
    t = jnp.sort(t, axis=1)
    return t


# Precompute sample of params, ICs and times
key = jax.random.PRNGKey(0)
W = jnp.array([[1.0]]) # Homogeneous system
print(f"Precomputing dataset with {N_traj} trajectories...")
key, sub = jrandom.split(key)
mu_b, x0_b = sample_batch(sub, N_traj)
t_b = sample_times(sub, N_traj, N_time) #Sample N_traj different time-grids (one per trajectory). This ensures the model doesn't just learn a single time-sequence

# Precompute supervised trajectories
u_b = vmap(lambda m, x, t: integrate(m, x, t, W))(mu_b, x0_b, t_b)
u_base_b = vmap(lambda x, t: vmap(lambda tt: baseline_solution(x, tt))(t))(x0_b, t_b)
residuals = u_b - u_base_b # Residuals with respect to baseline solution
print(f"Data generation done! Shape of u_b: {u_b.shape}")

# Compute the scale of the residuals (we use the max absolute deviation to keep NN outputs within [-1, 1])
u_scale_res = jnp.max(jnp.abs(residuals), axis=(0, 1))
u_scale_res = jnp.where(u_scale_res == 0, 1.0, u_scale_res)
print(f"Residual Scale (Theta, Sigma): {u_scale_res}")

# Model initialization
model = MultiInputDeepONet(
    mu_dim=mu_dim,
    x0_dim=x0_dim,
    latent_dim=latent_dim,
    basis_size=basis_size,
    system_dimension=system_dimension,
    t_max=t_max,
    mu_active_idx=[0],
    mu_min=jnp.array(eps_min),
    mu_max=jnp.array(eps_max),
    u_scale=u_scale_res,
    key=key
    )


#3. TRAINING

# Optimization setup
optimizer = optax.chain(optax.clip_by_global_norm(1.0), optax.adam(learning_rate=learning_rate))
opt_state_phase = optimizer.init(eqx.filter(model, eqx.is_inexact_array))

# Training step function
@eqx.filter_jit
def make_step(model, opt_state, mu, x0, t, u_true, W, weights, lambda_ode):
    loss_val, grads = eqx.filter_value_and_grad(loss)(model, mu, x0, t, u_true, W, weights, lambda_ode) # Loss and its gradient
    updates, opt_state = optimizer.update(
        grads,
        opt_state,
        eqx.filter(model, eqx.is_array)
    )
    model = eqx.apply_updates(model, updates) # Update DeepONet params
    return model, opt_state, loss_val

# Term multiplying ODE loss schedule
def lambda_ode_schedule(step, t0=2000, tau=500):
    lambda_min = 1e-6
    lambda_max = 1e-1
    s = jnn.sigmoid((step - t0) / tau)
    return lambda_min + (lambda_max - lambda_min) * s

# Training loop
print("Starting...")
save_dir = "/content/drive/MyDrive/deeponet_runs/PhaseAmplitude"
os.makedirs(save_dir, exist_ok=True)

loss_hist = []
N_batches = int(N_traj / Batch_size) # Number of considered batches per epoch
pbar = tqdm(range(N_epoch*N_batches), desc="Training PI-DeepONet")
for step in pbar: # We iterate over mini-batch weight updates
    key, sub_idx, sub_coll = jax.random.split(key, 3)

    # Supervised Batch
    idx = jax.random.randint(sub_idx, (N_sup,), 0, N_traj) # Sample indices for supervised batch . N_traj is the total number of precomputed trajectories, and we sample N_sup of them for this batch
    mu_sup = mu_b[idx]
    x0_sup = x0_b[idx]
    t_sup = t_b[idx]
    u_sup = u_b[idx]
    weights_sup = jnp.ones(N_sup) # Enable data loss

    # Collocation Batch (Freshly Sampled)
    mu_coll, x0_coll = sample_batch(sub_coll, N_coll)
    t_coll = sample_times(sub_coll, N_coll, N_time)
    u_coll = jnp.zeros((N_coll, N_time, 2)) # Placeholder (not used because weight=0)
    weights_coll = jnp.zeros(N_coll)    # Disable data loss

    #  Concatenate Batches
    mu = jnp.concatenate([mu_sup, mu_coll], axis=0)
    x0 = jnp.concatenate([x0_sup, x0_coll], axis=0)
    t = jnp.concatenate([t_sup, t_coll], axis=0)
    u = jnp.concatenate([u_sup, u_coll], axis=0)
    weights = jnp.concatenate([weights_sup, weights_coll], axis=0)

    # lambda_ode scheduling: Start with low weight on ODE loss
    lambda_ode = 0 # We disable ODE loss. To enable: lambda_ode_schedule(step)

    # Training step
    model, opt_state_phase, loss = make_step(
        model, opt_state_phase, mu, x0, t, u, W, weights, lambda_ode
    )

    loss_hist.append(float(loss))

    if step % 100 == 0:
        pbar.set_postfix({"loss": f"{loss:.4e}"})

    if step % 500 == 0:
        print(f"Step (mini-batch weight update) {step}, Loss: {loss:.4e}, Lambda_ODE: {lambda_ode:.2e}")
        debug = loss_debug(model, mu, x0, t, u, W, weights, lambda_ode)
        print(
            f"  Loss breakdown - Total: {debug[0]:.4e}, "
            f"Data: {debug[1]:.4e}, lambdaODE: {debug[2]:.4e}, "
            f"ODE: {debug[3]:.4e}"
        )


# Save trained model
eqx.tree_serialise_leaves(
    f"{save_dir}/model.eqx",
    model
)

# Moving average plot
window_size = 1000
moving_avg = jnp.convolve(
    jnp.array(loss_hist),
    jnp.ones(window_size) / window_size,
    mode='valid'
)

# Create the plot
plt.figure(figsize=(10, 6))
plt.semilogy(loss_hist, alpha=0.3, color='dodgerblue', label='Raw Batch Loss')
steps_ma = range(window_size - 1, len(loss_hist))
plt.semilogy(steps_ma, moving_avg, color='navy', linewidth=2, label=f'Moving Average (of the last {window_size} steps)')

plt.title("Training Loss Convergence")
plt.xlabel("Step (Mini-batch Weight Update)")
plt.ylabel("Mean Squared Error")
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend(loc='upper right')

# Save and render
plt.savefig(f"{save_dir}/loss_plot_with_ma.png", dpi=300)
plt.show()
```

```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import diffrax as dfx
import equinox as eqx

# 4. PREDICTION

# Load model
model = eqx.tree_deserialise_leaves(
    "/content/drive/MyDrive/deeponet_runs/PhaseAmplitude/model.eqx",
    #"/content/model.eqx",
    model
)

######################################
# Parameters, ICs and times to test!
W = jnp.array([[1.0]])
eps_test = 0.29 # [0,1]
theta_ic_test = 4.12 # [0,T/4]
sigma_ic_test = 0.58 # [-1,1]
t_test = jnp.linspace(0.0, T, 500)
######################################

mu = jnp.zeros((1,))
mu = mu.at[0].set(eps_test)
x0 = jnp.array([theta_ic_test, sigma_ic_test])
t = t_test

# Integrated trajectory (supervision)
u_true = integrate(mu, x0, t, W)
theta_true = u_true[:, 0]
sigma_true = u_true[:, 1]

# Predicted trajectory
u_p = jax.vmap(lambda tt: model(mu, x0, tt))(t)
theta_p = u_p[:, 0]
sigma_p = u_p[:, 1]

# Plot theta trajectory comparison
plt.figure(figsize=(8,5))
plt.plot(t, theta_true, label=f"ODE (Diffrax) [eps={eps_test:.2f}, x0=[{theta_ic_test:.2f}, {sigma_ic_test:.2f}]]", linewidth=3)
plt.plot(t, theta_p, "--", label="DeepONet")
plt.xlabel("t")
plt.ylabel("theta(t)")
plt.title("Trajectory comparison")
plt.legend()
plt.grid(True)
plt.savefig(f"{save_dir}/traj_compar_theta.png")
plt.show()

# Plot sigma trajectory comparison
plt.figure(figsize=(8,5))
plt.plot(t, sigma_true, label=f"ODE (Diffrax) [eps={eps_test:.2f}, x0=[{theta_ic_test:.2f}, {sigma_ic_test:.2f}]]", linewidth=3)
plt.plot(t, sigma_p, "--", label="DeepONet")
plt.xlabel("t")
plt.ylabel("sigma(t)")
plt.title("Trajectory comparison")
plt.legend()
plt.grid(True)
plt.savefig(f"{save_dir}/traj_compar_sigma.png")
plt.show()


# 5. INVERSE PROBLEM

print("\n Starting Inverse Problem Solver")

# Real trajectories from which we want to infer epsilon, theta0 and sigma0
true_mu = mu.at[0].set(eps_test)
true_x0 = jnp.array([theta_ic_test, sigma_ic_test])
t_observed = t_test

# Generate "observed" trajectory (without noise)
u_clean = integrate(true_mu, true_x0, t_observed, W)

# Add Gaussian measurement noise
key, noise_key = jax.random.split(key)
noise_level = 0.1
u_noisy = u_clean + noise_level * jax.random.normal(noise_key, u_clean.shape)

# Initialize parallel search
N_paths = 320
key, k1, k2 = jax.random.split(key, 3)

# Sample uniform distributions inside domain bounds
initial_mus = jax.random.uniform(k1, (N_paths, 1), minval=eps_min+1e-5, maxval=eps_max)
initial_x0s = jax.random.uniform(k2, (N_paths, 2), minval=ic_min, maxval=ic_max)

# IP loss
def single_inverse_loss(mu, x0, model, t_obs, u_obs_target):
    u_pred = jax.vmap(lambda tt: model(mu, x0, tt))(t_obs)
    scaled_error = (u_pred - u_obs_target) / model.u_scale
    return jnp.mean(scaled_error**2)

# Optimization setupp
inference_optimizer = optax.adam(learning_rate=5e-2)
init_params_flat = jnp.concatenate([initial_mus, initial_x0s], axis=1) # (N_paths, 3) shape
opt_state_inf = jax.vmap(inference_optimizer.init)(init_params_flat)

@jax.jit
def parallel_inference_step(mus, x0s, opt_state, t_obs, u_obs_target):
    # Compute losses across all searches
    losses = jax.vmap(lambda m, x: single_inverse_loss(m, x, model, t_obs, u_obs_target))(mus, x0s)

    # Compute gradients
    def get_single_grad(m, x):
        loss_val, (g_m, g_x) = jax.value_and_grad(lambda m_in, x_in: single_inverse_loss(m_in, x_in, model, t_obs, u_obs_target),argnums=(0, 1))(m, x)
        return loss_val, g_m, g_x

    _, grads_mu, grads_x0 = jax.vmap(get_single_grad)(mus, x0s)
    params_flat = jnp.concatenate([mus, x0s], axis=1)
    grads_flat = jnp.concatenate([grads_mu, grads_x0], axis=1)

    # Update params and IC
    updates, opt_state = jax.vmap(inference_optimizer.update)(grads_flat, opt_state, params_flat)
    new_params_flat = jax.vmap(optax.apply_updates)(params_flat, updates)
    new_mus = new_params_flat[:, :1]
    new_x0s = new_params_flat[:, 1:]

    # Restrict domain margins to prevent optimization paths collapsing to zero
    constrained_mus = jnp.clip(new_mus, eps_min+1e-5, eps_max)
    constrained_x0s = jnp.clip(new_x0s, ic_min, ic_max)

    return constrained_mus, constrained_x0s, opt_state, losses


# Inference routine
n_inference_steps = 1000
print("Optimizing parameter space for parallel paths...")

current_mus = initial_mus
current_x0s = initial_x0s

for step in range(n_inference_steps + 1):
    current_mus, current_x0s, opt_state_inf, individual_losses = parallel_inference_step(
        current_mus, current_x0s, opt_state_inf, t_observed, u_noisy
    )

    if step % 200 == 0:
        best_run_idx = jnp.argmin(individual_losses)
        current_best_loss = individual_losses[best_run_idx]
        current_best_eps = current_mus[best_run_idx, 0]
        print(f"Step {step:4d} | Global Best Loss: {current_best_loss:.4e} | Inferred eps: {current_best_eps:.4f}")

# Select global minimum from the paraller paths
best_idx = jnp.argmin(individual_losses)
final_recovered_mu = current_mus[best_idx]
final_recovered_x0 = current_x0s[best_idx]

print("\n Parallel Optimization Complete")
print(f"Ground Truth values -> eps: {true_mu[0]:.4f} | x0: {true_x0}")
print(f"Best Recovered Run  -> eps: {final_recovered_mu[0]:.4f} | x0: {final_recovered_x0}")


# Plot recovered inferred trajectories
u_recovered = jax.vmap(lambda tt: model(final_recovered_mu, final_recovered_x0, tt))(t_observed)

fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Plot theta(t)
ax[0].plot(t_observed, u_noisy[:, 0], 'kx', alpha=0.4,
           label=f"Observed Trajectory Points (Noise σ={noise_level*100:.1f}%)")
ax[0].plot(t_observed, u_recovered[:, 0], 'r-', linewidth=2,
           label=f"DeepONet Recovered\nTrue $\epsilon$={true_mu[0]:.3f} vs Pred={final_recovered_mu[0]:.3f}\nTrue $\\theta_0$={true_x0[0]:.3f} vs Pred={final_recovered_x0[0]:.3f}")
ax[0].set_title(r"Parameter Inference ($\theta$ Phase)")
ax[0].set_xlabel("t")
ax[0].set_ylabel("theta(t)")
ax[0].legend(loc="upper left")
ax[0].grid(True)

# Plot sigma(t)
ax[1].plot(t_observed, u_noisy[:, 1], 'kx', alpha=0.4,
           label=f"Observed Trajectory Points (Noise σ={noise_level*100:.1f}%)")
ax[1].plot(t_observed, u_recovered[:, 1], 'b-', linewidth=2,
           label=f"DeepONet Recovered\nTrue $\epsilon$={true_mu[0]:.3f} vs Pred={final_recovered_mu[0]:.3f}\nTrue $\\sigma_0$={true_x0[1]:.3f} vs Pred={final_recovered_x0[1]:.3f}")
ax[1].set_title(r"Parameter Inference ($\sigma$ Amplitude)")
ax[1].set_xlabel("t")
ax[1].set_ylabel("sigma(t)")
ax[1].legend(loc="upper left")
ax[1].grid(True)

plt.tight_layout()
plt.savefig(f"{save_dir}/traj_eps_IC_prediction.png")
plt.show()
```
